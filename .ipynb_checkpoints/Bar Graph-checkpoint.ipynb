```python
%matplotlib notebook
from modsim import *
import matplotlib.pyplot as plt
import numpy as np
```

```python
newfig()
ind = np.arange(1, 5)
height = [89.525894, 104.143188, 94.919174, 83.628419]
figure = plt.subplot()
plt.show(block=True)
ab, cd, ef = plt.bar(ind, height)
ab.set_facecolor('r')
cd.set_facecolor('b')
ef.set_facecolor('g')
figure.set_xticks(ind)
figure.set_xticklabels(['X increase', 'X decrese', 'Y increase', 'Y decrease'])
decorate(title='2% Variation in Initial Conditions',
         xlabel='Condition Changed',
         ylabel='Time of Initial Divergence (seconds)')
```
