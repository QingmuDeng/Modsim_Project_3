```python
%matplotlib notebook
from modsim import *
import matplotlib.pyplot as plt
import numpy as np
```

```python
newfig()
ind = [1, 2, 4, 5, 7, 8]
height = [89.525894, 104.143188, 94.919174, 83.628419, 28.396240, 18.350748] #, 76.907172, 90.864649]
figure = plt.subplot()
plt.show(block=True)
x_high, y_high, vy_high  = plt.bar(ind[::2], height[::2], width=1, label = 'Increase')
x_low, y_low, vy_low  = plt.bar(ind[1::2], height[1::2], width=1, label = 'Decrease')
x_high.set_facecolor('r')
x_low.set_facecolor('b')
y_high.set_facecolor('r')
y_low.set_facecolor('b')
vy_high.set_facecolor('r')
vy_low.set_facecolor('b')
# vx_high.set_facecolor('r')
# vx_low.set_facecolor('b')
figure.set_xticks([1.5, 4.5, 7.5])#, 10.5])
figure.set_xticklabels(['X', 'Y', 'VY'])#, 'VX'])
decorate(title='2% Variation in Initial Conditions',
         xlabel='Condition Changed',
         ylabel='Time of Initial Divergence (seconds)')
#plt.legend(['Increase','Decrease'], loc='upper left')
#legend([y_high], ["Increase"])
plt.grid()
plt.show()
```
